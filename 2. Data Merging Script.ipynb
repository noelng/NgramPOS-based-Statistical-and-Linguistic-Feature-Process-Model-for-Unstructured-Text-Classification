```python
import pandas as pd
forum = pd.read_csv("forum_topglove.csv")
news = pd.read_csv("news_topglove.csv")
```

```python
news.drop(['Unnamed: 0'], axis = 1, inplace = True)
forum.drop(['Unnamed: 0'], axis = 1, inplace = True)
```

## News

```python
news['Date'] = news['Date'].apply(lambda x: x.replace("\n", ''))
news['News'] = news['News'].apply(lambda x: x.replace("\n", ''))

day = ['Monday, ', 'Tuesday, ', 'Wednesday, ', 'Thursday, ', 'Friday, ', 'Saturday, ', 'Sunday, ']
for y in day:
    news['Date'] = news['Date'].apply(lambda x: x.replace(y, ''))
```

```python
news['Date'] = pd.to_datetime(news['Date'])
```

```python
news = news.sort_values(by="Date",ascending=True)
```

```python
news.columns = news.columns.str.replace('News', 'Text')
```

```python
news
```

## Forum

```python
forum['Date'] = forum['Date'].map(lambda x: str(x)[:-6])
forum = forum[~((forum['Date'] == '5 da') | (forum['Date'] == '2 mont') | (forum['Date'] == '3 wee') | (forum['Date'] == '1 d')
                | (forum['Date'] == '1 mon') | (forum['Date'] == '2 wee') | (forum['Date'] == '9 minut') 
                | (forum['Date'] == '1 we') | (forum['Date'] == '31 minut') | (forum['Date'] == '57 minut')
                | (forum['Date'] == '4 wee') | (forum['Date'] == '5 hou') | (forum['Date'] == '4 hou'))]
```

```python
forum['Date'].unique()
```

```python
forum['Date'] = pd.to_datetime(forum['Date'])
```

```python
forum = forum.sort_values(by="Date",ascending=True)
```

```python
forum
```

## Merge table

```python
df = pd.concat([news, forum])
```

```python
df
```

```python
df = df.sort_values(by="Date",ascending=True)
```

```python
df = df.groupby('Date', as_index=False).agg({"Text": "sum"})
```

```python
df.columns = df.columns.str.replace('Text', 'text')
```

```python
df
```

```python
df.to_csv("merged_Topglove.csv")
```
