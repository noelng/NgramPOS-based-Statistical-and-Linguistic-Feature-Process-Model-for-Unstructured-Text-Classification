```python
import pandas as pd
```

```python
df = pd.read_csv("merged_Topglove.csv")
stock = pd.read_csv("stock Topglove 7113.KL.csv")
```

```python
merge_df = pd.merge(stock, df, on='Date', how='left')
merge_df
```

```python
merge_df['Difference'] = ((merge_df['Close'] - merge_df['Open'])/merge_df['Open'])*100
```

```python
def labelling(data):
    if data > 0:
        label = 'Positive'
    elif data < 0:
        label = 'Negative'
    elif data == 0:
        label = 'Neutral'
    else:
        label = 'NA'
    return label
merge_df['label'] = merge_df['Difference'].apply(lambda x: labelling(x))
```
